```python
def tsp(*frags):
    # Converting strings into a list for easy calling
    fragments = list(frags)

    def join_frags(frag1, frag2): #function to join fragments
        max_overlap = 0
        min_len = min(len(frag1), len(frag2))
        for i in range(1, min_len + 1):
            if frag1[-i:] == frag2[:i]:
                max_overlap = i
        return frag1 + frag2[max_overlap:]

    def overlap(frag1, frag2): #Weight defining function based on overlap
        max_overlap = 0
        for i in range(1, min(len(frag1), len(frag2)) + 1):
            if frag1[-i:] == frag2[:i]:
                max_overlap = i
        return max_overlap

    # First fragment to start with for a Heuristic approach
    current_frag = fragments[0]
    # Remaining fragments to iterate upon
    remaining_frags = fragments[1:]
    while remaining_frags:
        Overlaps = [] #Provides weights in the form of overlaps
        for frag in remaining_frags:
            Overlaps.append(overlap(current_frag, frag))

        max_val = max(Overlaps) #Depicts the maximum weight amongst all edges in this iteration
        if max_val > 0:
          target = Overlaps.index(max_val) #Picks the target edge with max weight

          # Salesperson traveling accross that edge
          current_frag = join_frags(current_frag, remaining_frags[target])

          # Remove the fragment that has been joined
          remaining_frags.pop(target)

          # Print current_frag after each iteration for clarity
          print(f"Current fragment after joining: {current_frag}")

        elif True: #Repeating steps for fragments before current fragment

          Overlaps_rfrg = [] #For remaining fragments, list of weights

          for fragg in remaining_frags:
            Overlaps_rfrg.append(overlap(fragg,current_frag))

          maxi_val = max(Overlaps_rfrg)
          if maxi_val > 0:
            target = Overlaps_rfrg.index(maxi_val)

            #Joining rest of the fragments
            current_frag = join_frags(remaining_frags[target],current_frag)

            remaining_frags.pop(target)
            print(f"Current fragment after joining: {current_frag}")
          else:
            break

    return print(f'The final sequence is {current_frag}')
```
