```python
import itertools

def tsp_bnb(*frags):
    # Generate all permutations of the input arguments
    permutations = list(itertools.permutations(frags))

    def overlap(frag1, frag2):
        # Helper function to find the maximum overlap between two fragments
        max_overlap = 0
        min_len = min(len(frag1), len(frag2))
        for i in range(1, min_len + 1):
            if frag1[-i:] == frag2[:i]:
                max_overlap = i
        return max_overlap

    def join_frags(*fragments):  # Function to join multiple fragments
    # Initialize the result with the first fragment
     frag_1 = fragments[0]

    # Iterate through the remaining fragments and merge them
     for i in range(1, len(fragments)):
        frag_2 = fragments[i]
        max_overlap = overlap(frag_1, frag_2)
        frag_1 += frag_2[max_overlap:]

     return frag_1

    def all_overlaps(*fragments):  # Function to join multiple fragments
     # Initialize the result with the first fragment
     frag_1 = fragments[0]

     # Iterate through the remaining fragments and merge them
     SUM = []
     for i in range(1, len(fragments)):
      frag_2 = fragments[i]
      max_overlap = overlap(frag_1, frag_2)
      frag_1 += frag_2[max_overlap:]
      SUM.append(max_overlap)

     return sum(SUM)

    Weights = []
    for p in range(len(permutations)):
     Weights.append(all_overlaps(*permutations[p]))
    max_weight = max(Weights)
    My_index = Weights.index(max_weight)
    target = permutations[My_index]
    return join_frags(*target)
    # Convert tuples to lists
    #return [list(p) for p in permutations]
```

```python
a = ('ACTAGCATCGACTC','GATCTAC','CATTATCTCATCAGGCAT','TCTTATTCT','CTGCTTCATT')
```

```python
tsp_bnb(*a)
```
